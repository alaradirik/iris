```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
# Read data into pandas dataframe
df = pd.read_table('../iris.data', sep=",", skiprows=0)
df.columns = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species']
print(df.head())
```

```python
# Create features and labels arrays
target = np.array(df['Species'])
X = np.array(df.drop(['Species'], axis=1))
```

```python
# Plot correlations between features
# Graphics in SVG format are more sharp and legible
%config InlineBackend.figure_format = 'svg' 
sns.pairplot(df, hue='Species');
```

```python
# Split the dataset into training and testing groups
x_train, x_test, y_train, y_test = train_test_split(X, target, test_size=0.1, stratify=target)
```

```python
# Implement linear SVM model
model_linear = LinearSVC()
model_linear.fit(x_train, y_train.ravel())
```

```python
y_pred_linear = model_linear.predict(x_test)
print(model_linear.score(x_test,y_test.ravel()))
```

```python
# Create confusion matrix
cm_linear = confusion_matrix(y_test, y_pred_linear)
df_cm_linear = pd.DataFrame(cm_linear, index = [i for i in np.unique(target)],
                  columns = [i for i in np.unique(target)])
plt.figure(figsize = (5,5))
sns.heatmap(df_cm_linear, annot=True)
```

```python
#  Implement non-linear SVM model
model = SVC()
model.fit(x_train, y_train.ravel())
```

```python
y_pred = model.predict(x_test)
print(model.score(x_test,y_test.ravel()))
```
